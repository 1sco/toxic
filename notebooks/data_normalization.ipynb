```python
import pandas as pd
from pathlib import Path
import re
from functools import partial
```

```python
data_path = Path('../data')
train_data = pd.read_csv(Path(data_path, 'train.csv'))
test_data = pd.read_csv(Path(data_path, 'test.csv'))
```

```python
print(train_data.shape)
print(test_data.shape)
```

```python
test_data.head()
```

```python
pd.set_option('max_colwidth', -1)
```

```python
train_data['comment_text'].str.split().apply(lambda x: len(x)).describe()
```

```python
def remove_ip_address(text):
    ip_address_regex = r"[0-9]{1,3}\.[0-9]{1,3}\.[0-9]{1,3}\.[0-9]{1,3}"
    return re.sub(ip_address_regex, "", text)
    
def replace_num(text, symbol='#'):
    return re.sub(r"[0-9]", symbol, text)
```

```python
normalized_train_data = train_data.copy()
```

```python
normalized_train_data.head()
```

```python
normalized_train_data['comment_text'] = normalized_train_data['comment_text'].apply(remove_ip_address)
normalized_train_data['comment_text'] = normalized_train_data['comment_text'].apply(replace_num)
```

```python
normalized_train_data['comment_text'].str.split().apply(lambda x: len(x)).describe()
```

```python
normalized_train_data.to_csv(Path(data_path, 'normalized_train.csv'), index=False)
```

```python
normalized_test_data = test_data.copy()
normalized_test_data['comment_text'] = normalized_test_data['comment_text'].apply(remove_ip_address)
normalized_test_data['comment_text'] = normalized_test_data['comment_text'].apply(replace_num)
```

```python
normalized_test_data.to_csv(Path(data_path, 'normalized_test.csv'), index=False)
```
